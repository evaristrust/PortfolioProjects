```python
#Loop in my dataset and change all data with "object" data types into "Float" for correlation analysis
df_object = df_new

for col_name in df_object:
    if df_object[col_name].dtype == 'object':
        df_object[col_name] = df_object[col_name].astype('float')

df_object.head()
```

```python
#Create a heatmap chart to see how each feature is correlated to one another 
corr_matrix = df_new.corr(method='pearson')
sns.heatmap(corr_matrix, annot=True)
plt.xlabel('Features name')
plt.ylabel('Features name')
plt.title('Correlation Matrix for numerical data')
plt.show()
```

```python
## In the above heatmap, I am trying to figure out some factors that would highly affect our Total revenues

# We can see that:
# 'No of Households','Households W/ LPG', 'Urbanized Households', 'Houses in Rural', 'Housesholds W Firewood'
# 'Monthly Charcoal Cost - per household', 'Annual Charcoal - per household', 'Annual Savings - per household- Jikokoa'
## ARE AMONG THE FACTORS WE CAN NOW RELY ON WHILE DETERMINING A COUNTRY WE CAN TRY TO INVEST OUR BUSINESS MODEL MORE
# Because they are highly negatively or positively correlated with Total Revenues
```
