Data wrangling is known as the process of converting data from the initial format to a format that may be better for analysis.


```python
#import required libraries

import pandas as pd 
import numpy as np 
import matplotlib.pyplot as plt 
%matplotlib inline
import matplotlib.pyplot as plt
```

```python
# Going to wrangle automobile data 
# The file I will be reading 
file = 'https://cf-courses-data.s3.us.cloud-object-storage.appdomain.cloud/IBMDeveloperSkillsNetwork-DA0101EN-SkillsNetwork/labs/Data%20files/auto.csv'
```

```python
#Read the csv file 

pd.read_csv(file)
```

```python
#Now see the table does not contain headers
#Our columns now are needed to establish a clear table to analyze

columns = ["symboling","normalized-losses","manufacturer","fuel-type","aspiration", "num-of-doors","body-style",
         "drive-wheels","engine-location","wheel-base", "length","width","height","curb-weight","engine-type",
         "num-of-cylinders", "engine-size","fuel-system","bore","stroke","compression-ratio","horsepower",
         "peak-rpm","city-mpg","highway-mpg","price"]
```

```python
# Now read my data including the headers 
```

```python
df = pd.read_csv(file, names=columns)
```

```python
df
```

My data has missing information.
Going now to identify them and replace them where necessary

```python
#replace question mark with nan first

df.replace('?', np.nan, inplace=True)
```

```python
#check the top five rows now 
df.head()
```

```python
#alright now question mark (?) has been replace by nan
```

```python
#check how many null values in each columns

for col in df.columns:
    missing_data = np.sum(df[col].isnull())
    print('{} -- {}'.format(col, missing_data))
```

Based on the summary above, columns containing missing data are:

<ol>
    <li>"normalized-losses": 41</li>
    <li>"num-of-doors": 2</li>
    <li>"bore": 4</li>
    <li>"stroke" : 4</li>
    <li>"horsepower": 2</li>
    <li>"peak-rpm": 2</li>
    <li>"price": 4</li>
</ol>


```python
#going to replace null values by mean for columns 

# Normalized-losses, bore, stroke, horsepower, peak-rpm
```

```python
#Normalized-losses: 
avg_normalized_loss = df['normalized-losses'].astype('float').mean()
df['normalized-losses'].replace(np.nan, avg_normalized_loss, inplace=True)
```

```python
#Bore: 
avg_bore = df['bore'].astype('float').mean()
df['bore'].replace(np.nan, avg_bore, inplace=True)
```

```python
#Stroke: 
avg_stroke = df['stroke'].astype('float').mean()
df['stroke'].replace(np.nan, avg_stroke, inplace=True)
```

```python
#Horsepower: 
avg_horsepower = df['horsepower'].astype('float').mean()
df['horsepower'].replace(np.nan, avg_horsepower, inplace=True)
```

```python
#Peak-rpm: 
avg_peak = df['peak-rpm'].astype('float').mean()
df['peak-rpm'].replace(np.nan, avg_peak, inplace=True)
```

```python
#now check the missing values

df.isnull().sum()
```

```python
#now that we are remaining with price and num-of-doors
```

```python
#Num-of-doors: check values and their counts 

df['num-of-doors'].value_counts()
```

```python
#Now that cars with four doors are so many.. we can replace the null values with "four"
```

```python
#we can do it in many ways
#like, df['num-of-doors'].replace(np.nan, 'four', inplace=True)
#or check the idmax which I am going to use here
```

```python
max_type = df['num-of-doors'].value_counts().idxmax()
```

```python
#Replace null values in num-of-doors with max_type

df['num-of-doors'].replace(np.nan, max_type, inplace=True)
```

```python
#For the sake of my analysis, I don't need a row with an empty price value
#I am going to drop them 
```

```python
df.dropna(subset=['price'], axis=0, inplace=True)
```

```python
#now reset index, because we dropped rows
df.reset_index(drop=True, inplace=True)
```

```python
df.head()
```

```python
#Verify if I now have an empty free data 

df.isnull().sum()
```

Great!!! Now I am good to go

```python
#After identifying and correcting null values,
#We going to check the data types of each column
```

```python
df.dtypes
```

```python
#Going to change
#Normalized-losses and horsepower to 'int'
#bore, stroke, price, and peak-rpm to float
```

```python
df[['bore', 'stroke', 'price', 'peak-rpm']] = df[['bore', 'stroke', 'price', 'peak-rpm']].astype('float')
df[['normalized-losses', 'horsepower']] = df[['normalized-losses', 'horsepower']].astype('int64')
```

```python
df.dtypes
```

NOW I AM GOOD TO GO WITH THE RIGHT DATA TYPES

```python
#Check column 'city-mpg' and 'highway-mpg' 
#mpg: mile per gallon is not common
#we can standardize it to liters per 100km
```

```python
#Now that L/100km = 235/mpg
```

```python
df['city-l/100km'] = 235 / df['city-mpg']
df['highway-l/100km'] = 235 / df['highway-mpg']
```

```python
df.head()
```

```python
#Let's now get rid of city-mpg and highway-mpg since we no longer need them 
df.drop(['city-mpg', 'highway-mpg'], axis=1, inplace=True)
```

```python
df.head()
```

```python
#Now we can also normalize columns 
#length, width, and height 
```

```python
#Normalization means transforming values of several variables into a small range!
#So their values range from 0 to 1
```

```python
#One way to do it is to divide each value by the maximum in the column
```

```python
df['length'] = df['length']/df['length'].max()
df['width'] = df['width']/df['width'].max()
df['height'] = df['height'] / df['height'].max()
```

```python
df[['length', 'width', 'height']].head()
```

GREAT! we keep going

```python
#Look at the column, Horsepower! 
#Going to do BINNING for this column
```

```python
#Binning is the process of transforming continuous numerical variables such as those of horsepower into categorical bins for grouped analysis 
```

```python
#What do we want now? 
#Want to categorize horspower data into LOW, MEDIUM, AND HIGH
```

```python
#SINCE I want to get three bins (Low, medium, and high),
# dividers should be 4
```

```python
bins = np.linspace(min(df['horsepower']), max(df['horsepower']),4)
bins
```

```python
groups = ['LOW', 'MEDIUM', 'HIGH']
```

```python
df['horsepower-binned'] = pd.cut(df['horsepower'], bins=bins, labels=groups,ordered=True, include_lowest=True)
```

```python
#Now see counts by category 
groups_counts = df['horsepower-binned'].value_counts()
groups_counts
```

```python
#Going to plot using matplotlib.pyplot library

plt.bar(groups, groups_counts)
plt.xlabel('Horsepower')
plt.ylabel('Counts')
plt.title('Horspower Bins')
plt.show()
```

```python
df['fuel-type'].value_counts()
```

```python
df['aspiration'].value_counts()
```

```python
#These are also categorical columns
#I am going to get dummy variables or indicator variable from them
```

```python
#Dummy variable is a numerical varible used to label categories 
```

```python
fuel_type_dummies = pd.get_dummies(df['fuel-type'])
fuel_type_dummies
```

```python
aspiration_dummies = pd.get_dummies(df['aspiration'])
aspiration_dummies
```

```python
#Just for more clarity, let's rename them 
```

```python
fuel_type_dummies.rename(columns={'gas': 'fuel-type-gas', 'diesel': 'fuel-type-diesel'}, inplace=True)
fuel_type_dummies.head()
```

```python
aspiration_dummies.rename(columns={'std': 'aspiration-std', 'turbo': 'aspiration-turbo'}, inplace=True)
aspiration_dummies.head()
```

```python
#Let's now merge dummies to our original dataset

df = pd.concat([df, fuel_type_dummies, aspiration_dummies], axis=1)
```

```python
#Now let's drop aspiration and fuel-type columns
```

```python
df.columns
```

```python
df.drop(['fuel-type', 'aspiration'], axis=1, inplace=True)
```
